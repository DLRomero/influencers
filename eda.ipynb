```python
def miles(x):
    letra_= re.findall('[A-Z]', str(x))
    numeros= re.findall('\d+\.*\d*', str(x))
    
    try:   
        if letra_[0]== 'K':
            return float(numeros[0])* 1000
        elif letra_[0]== 'M':
            return float(numeros[0])* 1000000
        
    except: 
            if len(numeros)== 1:
                return float(numeros[0])
            else:
                return np.nan
               
```

```python
yb.columns
```

```python
yb['avg likes']= yb['avg likes'].apply(miles)
```

```python
yb['avg views']= yb['avg views'].apply(miles)
```

```python
yb['Subscribers']= yb['Subscribers'].apply(miles)
```

```python
numericas= yb.select_dtypes(np.number)
numericas.head()
```

```python
imputerKNN = KNNImputer(n_neighbors=5)
imputerKNN.fit(numericas)
```

```python
numericas_knn= imputerKNN.transform(numericas)
```

```python
df_knn_imputer = pd.DataFrame(numericas_knn, columns = numericas.columns)
```

```python
df_knn_imputer.head()
```

```python

columnas_knn = df_knn_imputer.columns
```

```python
yb.drop(columnas_knn, axis = 1, inplace = True)
```

```python
yb[columnas_knn] = numericas_knn
```

```python
yb.isnull().sum()
```

```python
yb.to_csv('./Datos/youtube_limpio.csv')
```
